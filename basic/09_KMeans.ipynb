# 데이터들을 비슷한 속성끼리 분류

## K-Means 군집화 맛보기 : 인위적으로 만든 데이터셋

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_url = 'https://raw.githubusercontent.com/snkn1959/data_source/main/example_cluster.csv'
data = pd.read_csv(file_url)
```

```python
data
```

```python
sns.scatterplot(x='var_1', y='var_2', data=data) # 산점도로 데이터 분포 확인
```

```python
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=3, random_state=100) # 산점도를 통해 3개의 군집을 확인
kmeans_model.fit(data)
```

```python
kmeans_model.predict(data) # 예측
```

```python
data['label'] = kmeans_model.predict(data) # 예측 값을 label 변수 생성하여 저장
data
```

```python
sns.scatterplot(x='var_1', 
                y='var_2',
                data=data,
                hue='label', 
                palette='rainbow')
```

```python
# 이너셔 값 확인
# inertia : 각 그룹에서의 중심과 각 그룹에 해당하는 데이터 간의 거리에 대한 합
kmeans_model.inertia_ 
```

```python
temp_model = KMeans(n_clusters=500, random_state=100)
temp_model.fit(data)
temp_model.inertia_
```

```python
# 엘보우 기법을 통해 적절한 K 값 찾기
distance = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(data)
    distance.append(k_model.inertia_)
```

```python
distance
```

```python
sns.lineplot(x=range(2,10), y=distance) # 엘보우 그래프
```

## 데이터 확인하기 : 고객 데이터셋

```python
file_url = 'https://raw.githubusercontent.com/snkn1959/data_source/main/customer.csv'
customer = pd.read_csv(file_url)
customer.head()
```

```python
customer['cc_num'].nunique() # 고객 고유값 확인
```

```python
customer['category'].nunique() # 카테고리 고유값 확인
```

## 전처리 : 피처 엔지니어링

```python
customer_dummy = pd.get_dummies(customer, columns=['category'])
```

```python
customer_dummy.head()
```

```python
cat_list = customer_dummy.columns[2:]
for i in cat_list:
    customer_dummy[i] = customer_dummy[i] * customer_dummy['amt']
```

```python
customer_dummy
```

```python
customer_agg = customer_dummy.groupby('cc_num').sum() # cc_num별 총 금액
customer_agg.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(customer_agg)
scaled_df = pd.DataFrame(scaled, columns=customer_agg.columns, index=customer_agg.index)
```

```python
scaled_df.head()
```

## 고객 데이터 모델링 및 실루엣 개수

```python
distance = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaled_df)
    labels = k_model.predict(scaled_df)
    distance.append(k_model.inertia_)
    
sns.lineplot(x=range(2,10), y=distance)
```

```python
# 엘보우 그래프를 통해 K 값을 정하지 못할 경우 실루엣(silhuette) 기법을 통해 확인
from sklearn.metrics import silhouette_score
```

```python
silhouette = []
for k in range(2,10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaled_df)
    labels = k_model.predict(scaled_df)
    silhouette.append(silhouette_score(scaled_df, labels))
    
sns.lineplot(x=range(2,10), y=silhouette)
```

## 최종 예측 모델 및 결과 해석

```python
k_model = KMeans(n_clusters=4)
k_model.fit(scaled_df)
labels = k_model.predict(scaled_df)
```

```python
scaled_df['label'] = labels
scaled_df
```

```python
scaled_df_mean = scaled_df.groupby('label').mean() # label별 평균값
scaled_df_count = scaled_df.groupby('label').count()['category_travel'] # label별 등장 횟수
```

```python
# join할 경우 같은 컬럼이 존재하므로 컬럼명 변경
scaled_df_count = scaled_df_count.rename('count')
```

```python
scaled_df_all = scaled_df_mean.join(scaled_df_count)
scaled_df_all
```
